# Data Eksploration

```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set_style("whitegrid")
import matplotlib as mpl
import matplotlib.pyplot as plt

# library untuk klasifikasi Neural Network
from sklearn.neural_network import MLPClassifier

# library untuk klasifikasi SVM
from sklearn import svm
```

```python
ds = pd.read_csv('air_bnb.csv')
```

```python
ds
```

```python
feature_drop = ['minimum_nights','name','latitude','host_name','neighbourhood_group', 'neighbourhood',
               'id','host_id','room_type','longitude','last_review','price']
ds.drop(columns = feature_drop, inplace = True)
```

```python
ds
```

```python
ds.describe()
```

```python
ds.isna()
```

```python
df = ds.dropna()
```

```python
df
```

```python
mpl.rcParams['figure.dpi'] = 200
korelasi = df.select_dtypes(include = ['float64', 'int64']).iloc[:, :].corr()
plt.figure(figsize=(10, 10))
ax = sns.heatmap(korelasi, vmax=1, square=True)
# ax.set_xticklabels(rotation=30)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
```

```python
mpl.rcParams['figure.dpi'] = 300
plt.figure(figsize=(100, 100))
sns.pairplot(df[['availability_365', 'number_of_reviews','reviews_per_month','calculated_host_listings_count']] )
```

```python
data = np.array(df.drop('availability_365',1))
label = df['availability_365'].values
```

```python
data.shape
```

```python
data_train = data[:10000]
label_train = label[:10000]
dataset = data[10000:]
labelset = label[10000:]
```

# Menggunakan Neural Network

```python
# menggunakan multilayer method
mlp = MLPClassifier(solver='adam', alpha=1e-5, hidden_layer_sizes=(128,64,32,16,8,4,2), random_state = 1)
mlp.fit(data_train, label_train)
```

```python
prediksi = mlp.predict(data_train)
dataPred = zip(prediksi,labelset)
jml_label = labelset.shape[0]
for p, ls in enumerate(dataPred):
    if p == ls:
        jml_benar += 1
print('Akurasi', jml_benar/jml_label*100)
```

# Menggunakan SVM

```python
svm_clf = svm.SVC()
svm_clf.fit(data_train, label_train)
```

```python
prediksi = svm_clf.predict(data_train)
dataPred = zip(prediksi,labelset)
jml_label = labelset.shape[0]
for p, ls in enumerate(dataPred):
    if p == ls:
        jml_benar += 1
print('Akurasi', jml_benar/jml_label*100)
```
