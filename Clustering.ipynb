# Tahap Clustering

```python
# memvisualkan data dengan grafik
x = df['longitude'].values
y = df['latitude'].values
dataCluster = np.array([x,y]).T
plt.plot(x,y, 'b.')
plt.show
```

```python
# mendifine centroid
def defineCentroid(x,y,k):
    cx = np.random.uniform(min(x), max(x),k)
    cy = np.random.uniform(min(y), max(y),k)
    centroid = np.array([cx,cy]).T
    return centroid
```

```python
# menghitung euclidian distance
def euclidianDistance(x,y):
    return np.sqrt(np.sum((x-y)**2))
```

```python
# fungsi kMeans untuk update centroid
def kMeans(dataCluster, centroid, k):
    # melakukan update centroid dengan 50 iterasi
    for i in range(50):
        # mendifine cluster
        cluster = []
        for j in range(k):
            cluster.append([])
        # perulangan di dalam data Cluster
        for data in dataCluster:
            # menghitung distance data dan centroid
            distData = euclidianDistance(data, centroid)
            # variable menyimpan index
            d_idx = 0
            # perulangan selama i dan cen pada centroid
            for i, cen in enumerate(centroid):
                # menghitung distance data dan cen
                distC = euclidianDistance(data, cen)
                if distC < distData:
                    # mengubah nilai distance data dengan distance centroid
                    distData = distC
                    d_idx = i
            # mengisi cluster pada index saat itu dengan data
            cluster[d_idx].append(data)
        # perulangan untuk mengambil instansi dalam cluster
        for i, c in enumerate(cluster):
            if len(c) > 0:
                c = np.array(c)
                centroid[i] = c.mean(0)
    return centroid
```

```python
def clusterVisualize(dataCluster, centroid, k):
    # mendifine cluster
    cluster = []
    for j in range(k):
        cluster.append([])
    # perulangan di dalam data Cluster
    for data in dataCluster:
        # menghitung distance data dan centroid
        distData = euclidianDistance(data, centroid)
        d_idx = 0
        for i, cen in enumerate(centroid):
            # mengubah nilai distance data dengan distance centroid
            distC = euclidianDistance(data, cen)
            if distC < distData:
                # mengubah nilai distance data dengan distance centroid
                distData = distC
                d_idx = i
        # mengisi cluster pada index saat itu dengan data
        cluster[d_idx].append(data)
    for c in cluster:
        if len(c) > 0:
            c = np.array(c).T
            plt.scatter(c[0], c[1])
    plt.show
```

```python
k = 4
centroid = defineCentroid(x,y,k)
print('Centroid awal','\n', centroid)
centeroid = kMeans(dataCluster, centroid, k)
print('Centroid akhir','\n', centeroid)
clusterVisualize(dataCluster, centroid, k)
```

```python
k = 8
centroid = defineCentroid(x,y,k)
print('Centroid awal','\n', centroid)
centeroid = kMeans(dataCluster, centroid, k)
print('Centroid akhir','\n', centeroid)
clusterVisualize(dataCluster, centroid, k)
```
